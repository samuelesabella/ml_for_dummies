```
import numpy as np
import matplotlib.pyplot as plt
import random
```

```
values = np.random.normal(7, .5, 100)
plt.plot(values)
```

```
n_points = 1000


def make_inner(mu, sigma, num):
    x = np.random.normal(mu, sigma, num)
    y = np.random.normal(mu, sigma, num)
    return x, y

x_inner, y_inner = make_inner(0, 0.2, n_points)
plt.plot(x_inner, y_inner, 'o', markersize=2, c="C1")


def make_circle(mu=4., sigma=1., num=1000):
    r = np.random.normal(mu, sigma, num)
    phi = np.linspace(0,2.*np.pi, len(r))
    x = r * np.sin(phi)
    y = r * np.cos(phi)
    return x, y

x_circle, y_circle = make_circle(0.8, 0.05, n_points)
plt.plot(x_circle, y_circle, 'o', markersize=2)
```

```
import keras
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```
model = Sequential([
    Dense(7, input_shape=(2,), activation="relu"),
    Dense(5, input_shape=(2,), activation="relu"),
    Dense(2, activation="softmax"),
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```
def make_dataset(num):
    input_circle = np.array(list(zip(*make_circle(0.8, 0.05, num))))
    output_circle = np.repeat(np.array([[0, 1]]), num, axis=0)

    # input_inner = np.dstack(make_inner(0, 0.2, num))[0]
    input_inner = np.array(list(zip(*make_inner(0, 0.2, num))))
    output_inner = np.repeat(np.array([[1, 0]]), num, axis=0)

    input = np.concatenate((input_circle, input_inner), axis=0)
    output = np.concatenate((output_circle, output_inner), axis=0)

    return input, output

training_set = make_dataset(10000)
test_set = make_dataset(10000)
```

```
history = model.fit(training_set[0], training_set[1],
            epochs=40,
            batch_size=512,
            shuffle=True,
            validation_data=(test_set[0], test_set[1])
          ).history
```

```
plt.plot(history['loss'], color='b', label='loss')
plt.plot(history['val_loss'], color='g', 
linestyle='--', label='val_loss')
plt.legend()
plt.show()
```

```
tr = np.array(list(zip(*make_circle(0.8, 0.05, 1000))))

predicted = list(model.predict(tr))
predicted_colors = ['C1' if np.argmax(x)==0 else 'C0' for x in predicted]

for i in range(len(predicted)):
    plt.plot(tr[i,0], tr[i,1], 'o', markersize=2, c=predicted_colors[i])
```

```
predicted = model.predict(test_set[0])

circle = []
inner = []

for i in range(len(predicted)):
    if np.argmax(predicted[i]) == 0:
        circle.append(test_set[0][i])
    else:
        inner.append(test_set[0][i])

circle_x = [x[0] for x in circle]
circle_y = [x[1] for x in circle]
inner_x = [x[0] for x in inner]
inner_y = [x[1] for x in inner]


plt.plot(inner_x, inner_y, 'o', markersize=2)
plt.plot(circle_x,circle_y, 'o', markersize=2)
```
