```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Load an color image in grayscale
img = cv2.imread('Lenna.png',0) / 255.
```

```python
plt.imshow(img, cmap='gray') 
plt.axis(False);
```

# Convolving

```python
plt.figure(figsize=(20,10))
sobelx = cv2.Sobel(img,cv2.CV_64F, 1, 0, ksize=3)
plt.imshow(sobelx*2, cmap='gray', vmin=0, vmax=1) 
plt.axis(False);
```

# Implementing manual convolution

```python
from scipy import signal

sobelKernelX = np.array([[1,0,-1], [2,0,-2], [1,0,-1]])

def local_sobel(x,y, K):
    sub_k = img[x-1:x+2,y-1:y+2]
    Kfilter = np.flip(K, 0)
    Kfilter = np.flip(Kfilter, 1)
    return np.multiply(Kfilter, sub_k).sum()

def check_error(x,y,eps=1e-14):
    return abs(x-y)<1e-14

for i in range(5, 100):
    for j in range(5, 100):
        assert check_error(sobelx[i,j], local_sobel(i,j,sobelKernelX))
```
